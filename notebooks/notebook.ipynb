### 2.2.2 Age

```python
def age_category(age):
    if age >= 18 and age <= 40:
        return 'adult'
    elif age > 40 and age <= 50:
        return 'early_middle'
    elif age > 50 and age <= 60:
        return 'mid_middle'
    elif age > 60 and age <= 65:
        return 'late_middle'
    elif age > 65 and age <= 75:
        return 'older_adult'
    else:
        return 'older' # > 75
```

```python
train_derived['Age_Group'] = train_derived['Age'].apply(age_category)
train_derived['Age_Group'].value_counts()
```

### 2.2.3 Body

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return 'underweight'
    elif bmi >= 18.5 and bmi <= 24.9:
        return 'optimal'
    elif bmi > 24.9 and bmi <= 29.9:
        return 'overweight'
    elif bmi > 29.9 and bmi <= 34.9:
        return 'obese_1'
    elif bmi > 34.9 and bmi <= 39.9:
        return 'obese_2'
    else:
        return 'obese_3'
```

```python
train_derived['BMI'] = train_derived['Weight'] / (train_derived['Height'] / 100) ** 2
```

```python
train_derived['BMI_Category'] = train_derived['BMI'].apply(bmi_category)
```

```python
def features_transformation(data, kmeans_predictor, cluster_labels):
    data['Body_Temp_Incr'] = data['Body_Temp'] - 36.5
    data['HR_MAX'] = 220 - data['Age'] 
    data['HR_Incr'] = data['Heart_Rate'] - 60 
    data['HR_Rel'] = data['Heart_Rate'] / data['HR_MAX'] 
    data['Age_Group'] = data['Age'].apply(age_category)
    data['BMI'] = data['Weight'] / (data['Height'] / 100) ** 2
    data['BMI_Category'] = data['BMI'].apply(bmi_category)
    data['Weight_Effort'] = data['Weight'] * data['Duration'] * data['HR_Rel']
    data['Session_Intensity'] = data['Duration'] * data['HR_Incr'] * data['Body_Temp_Incr']
    data['Session_Performance'] = data['Duration'] * data['HR_Rel']
    data['Log_Weight_Effort'] = np.log1p(data['Weight'] * data['Duration'] * data['HR_Rel'])
    data['Log_Session_Intensity'] = np.log1p(data['Duration'] * data['HR_Incr'] * data['Body_Temp_Incr'])
    data['Log_Session_Performance'] = np.log1p(data['Duration'] * data['HR_Rel'])
    data['Session_Performance_Bins'] = pd.qcut(data['Session_Performance'], q=4, labels=['Low', 'Mid-Low', 'Mid-High', 'High'])

    from sklearn.cluster import KMeans
    from sklearn.preprocessing import StandardScaler

    X = data[['Duration', 'HR_Incr', 'Body_Temp_Incr']]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    kmeans = KMeans(n_clusters=4, random_state=42)
    data['Session_Type'] = kmeans.fit_predict(X_scaled)

    data['Session_Type'] = data['Session_Type'].map(cluster_labels)
    return data
```
